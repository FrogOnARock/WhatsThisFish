```python
import sqlalchemy.orm
from sqlalchemy import select, func

from whatsthatfish.src.database.models import SuccessfulUploads
from whatsthatfish.src.database.config import get_session_factory
from whatsthatfish.src.database.models import InatFilteredObservations
import re
from dotenv import load_dotenv


load_dotenv()
session_factory = get_session_factory()
with session_factory() as session:
    ids = session.execute(
        select(SuccessfulUploads.identifier).where(SuccessfulUploads.source=='inat')
    ).scalars().all()


with open("images.txt", 'r') as f:
    data = f.read()

files = [i.split("//")[-1].split(".")[0] for i in re.split(r"\n", data)]
files = set(files)
ids = set(ids)

length_ids = len(ids)
diff = ids - files
length_diff = len(diff)

print(list(diff)[:5])
print(f"Total different {length_diff}, percentage: {(length_diff / length_ids):.1%}")
```
